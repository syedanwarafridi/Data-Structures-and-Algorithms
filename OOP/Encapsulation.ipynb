# Encapsulation

- Instance Variable: Variable we create in Contstructor are called Instance Variable
- Because each Instance Variable value is different for different objects
- Encapsulation is mainly used for hiding the data which you dont want to give
- It protects your classes from accidental changes or deletions and promotes code reusability and maintainability.
- Python intrpreter will convert encapsulated memebers with in the name ("_Atm__pin", etc)

```python
class Atm:
    def __init__(self):
        self.__pin = ""
        self.__balance = 0
        self.__menu()
        
    def get_pin(self):
        return self.__pin
    
    def set_pin(self, new_pin):
        if type(new_pin) == str:
            self.__pin = new_pin
            print("Pin Changed Successfully")
        else:
            print("Not Allowed")
        
    def __menu(self):
        user_input = input("""
                           Hello, How would you like to proceed?
                           1. Enter 1 to create pin.
                           2. Enter 2 to Deposite.
                           3. Enter 3 to withdraw.
                           4. Enter 4 to check balance.
                           5. Enter 5 to Exit.
                           """)
        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.deposit()
        elif user_input == "3":
            self.withdraw()
        elif user_input == "4":
            self.check_balance()
        else:
            print("bye")
            
    def create_pin(self):
        self.__pin = input("Enter your pin: ")
        return f"Pin Created Successfully"
    
    def deposit(self):
        temp = input("Enter Your Pin Code: ")
        if temp == self.__pin:
            amount = int(input("Enter the amount: "))
            self.__balance += amount
            return f"Deposite Successfull"
        else:
            return f"Invalid pin"
    
    def withdraw(self):
        temp = input("Enter Your Pin Code: ")
        if temp == self.__pin:
            amount = int(input("Enter the amount: "))
            if amount <= self.__balance:
                self.__balance -= amount
                return f"Withdraw Successful"
            else:
                return f"Insufucient Balance"
        else:
            return f"Invalid Pin"
        
    def check_balance(self):
        temp = input("Enter Your Pin Code: ")
        if temp == self.__pin:
            return self.__balance
        else:
            return f"Invalid Pin"
```
